```
def rotate(img, count, outi, label, outl):
  # load the image
  img = cv2.imread(img)
  label = cv2.imread(label)
  # convert to numpy array
  data = img_to_array(img)
  datal = img_to_array(label)
  # expand dimension to one sample
  samples = expand_dims(data, 0)
  samplesl = expand_dims(datal, 0)
  # create image data augmentation generator
  datagen = ImageDataGenerator(rotation_range=45)
  # prepare iterator
  it = datagen.flow(samples, batch_size=1)
  itl = datagen.flow(samplesl, batch_size=1)
  # generate samples and plot
  for i in range(9):
    # define subplot
    #pyplot.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    batchl = itl.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    label = batchl[0].astype('uint8')
    cv2.imwrite(outi, image)
    cv2.imwrite(outl, label)
    # plot raw pixel data
    #pyplot.imshow(image)
  # show the figure
  #pyplot.show()
```

```
def verticle_shift(img, count, outi, label, outl):
	# load the image
  img = cv2.imread(img)
  label = cv2.imread(label)
  data = img_to_array(img)
  datal = img_to_array(label)
  # expand dimension to one sample
  samples = expand_dims(data, 0)
  samplesl = expand_dims(datal, 0)
  # create image data augmentation generator
  datagen = ImageDataGenerator(height_shift_range=0.5)
  # prepare iterator
  it = datagen.flow(samples, batch_size=1)
  itl = datagen.flow(samplesl, batch_size=1)
  # generate samples and plot
  for i in range(9):
    # define subplot
    #pyplot.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    batchl = itl.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    imagel = batchl[0].astype('uint8')
    cv2.imwrite(outi, image)
    cv2.imwrite(outl, imagel)
```

```
def flip(img, count, outi, label, outl):
  # load the image
  img = cv2.imread(img)
  label = cv2.imread(label)
  # convert to numpy array
  data = img_to_array(img)
  datal = img_to_array(label)
  # expand dimension to one sample
  samples = expand_dims(data, 0)
  samplesl = expand_dims(datal, 0)
  # create image data augmentation generator
  datagen = ImageDataGenerator(horizontal_flip=True)
  # prepare iterator
  it = datagen.flow(samples, batch_size=1)
  itl = datagen.flow(samplesl, batch_size=1)
  # generate samples and plot
  for i in range(9):
    # define subplot
    #pyplot.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    batchl = itl.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    label = batchl[0].astype('uint8')
    print(image.shape)
    cv2.imwrite(outi, image)
    cv2.imwrite(outl, label)
    # plot raw pixel data
    #pyplot.imshow(image)
  # show the figure
  #pyplot.show()
```

```
def color(img, count, out):
  img = cv2.imread(img)
  # convert to numpy array
  data = img_to_array(img)
  # expand dimension to one sample
  samples = expand_dims(data, 0)
  # create image data augmentation generator
  datagen = ImageDataGenerator(brightness_range=[0.2,1.0])
  # prepare iterator
  it = datagen.flow(samples, batch_size=1)
  # generate samples and plot
  for i in range(9):
    # define subplot
    #pyplot.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    cv2.imwrite(out, image)
    # plot raw pixel data
    #pyplot.imshow(image)
  # show the figure
  #pyplot.show()
```
